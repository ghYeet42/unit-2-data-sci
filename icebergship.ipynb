```python
import pandas as pd
import matplotlib.pyplot as pplt
from matplotlib import numpy as np
from collections import Counter

df = pd.read_csv("titanic.csv")

df.head()

# Counter(df["Pclass"])
```

```python
Counter(df["Pclass"])

Counter(df["Survived"])

Counter(df["Sex"])

Counter(df["Age"])

Counter(df["SibSp"])

Counter(df["Embarked"])
```

```python
mean_ages = df.groupby(["Sex", "Survived"])["Age"].mean()

print(mean_ages)

# Calculate mean ages based on 'Sex' and 'Survived'
mean_ages = df.groupby(["Sex", "Survived"])["Age"].mean()

# Specify names and corresponding values for the plot (average age of males and females who survived)
data = {
    "Male Survivors": mean_ages["male", 1],   # "male", 1 corresponds to male survivors 
    "Female Survivors": mean_ages["female", 1]  # "female", 1 corresponds to female survivors
}

# Extract names and values for the plot
names = list(data.keys())
values = list(data.values())

# Create a single subplot (1 row, 1 column)
fig, ax = pplt.subplots(figsize=(6, 4))

# Plotting the data as a bar graph
ax.bar(names, values, color='skyblue')

# Setting the title and labels for the plot
ax.set_title('Average Age of Male and Female Survivors')
ax.set_xlabel('Gender')
ax.set_ylabel('Mean Age')

# Display the bar graph
pplt.show()

# this graph shows that the average age of male and female survivors on the titanic, implying that if you were a male titanic passenger around the ages of 27 or 28 you were more likely to surviv, if you were a female titanic passenger around the ages of 27-28, you were more likely to survive.
```
